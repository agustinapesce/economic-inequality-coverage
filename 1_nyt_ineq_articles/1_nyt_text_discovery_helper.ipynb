```python
def extract_docx_from_zip(zip_path):
    """Extract .docx files from a ZIP archive."""
    docx_files = []
    
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        # Get list of all file names in the ZIP archive
        for file_name in zip_ref.namelist():
            #print(file_name)
            if file_name.endswith('.DOCX'):
                # Read the .docx file into memory
                with zip_ref.open(file_name) as docx_file:
                    docx_files.append(BytesIO(docx_file.read()))
    print(docx_files)
    return docx_files
```

```python
# List of words to search for
search_words = [
    "TODD:", "Good morning.", "Other Big Stories", "MORNING READS", "THE WALL STREET JOURNAL",
    "In the News", "Last Chance", "Dear Diary:", "BLOOMBERG NEWS",
    "War in Ukraine", "Close boyfriend loophole: Yes", "Quote of the Day",
    "Previews and Openings", "Off Broadway", "Off Off Broadway", "READ MORE:",
    "WHAT YOU’RE SAYING", "The tech flyaround", "The political flyaround",
    "What to Listen to", "Take our news quiz.", "RECOMMENDATIONS:"
]


def inner_titles_extraction(doc, folder_name):
    """Extract paragraphs with a single word matching the search words and their associated text."""
    matched_data = []

    # Combine all paragraphs into a single string of text
    full_text = "\n".join([paragraph.text.strip() for paragraph in doc.paragraphs])

    # Loop through all paragraphs in the document
    for paragraph in doc.paragraphs:
        # Split the paragraph text into words and check if it has only one word
        words_in_paragraph = paragraph.text.strip().split()
        if len(words_in_paragraph) == 1:  # Check if there is just one word
            word = words_in_paragraph[0]
            
            # Check if the single word matches any of the specified search words
            for search_word in search_words:
                if word.lower() == search_word.lower():  # Case-insensitive match
                    matched_data.append({
                        'full_text': full_text,  # The entire text of the document
                        'word_present': search_word,  # The word from search_words that matched
                        'folder_name': folder_name  # Folder name where the .docx file was extracted
                    })
                    break  # If a match is found, no need to check other words
    
    return matched_data

def inner_titles_extraction_zip_folders(folder_path):
    """Process all ZIP files in a folder and extract data from .docx files."""
    all_data = []

    # Get the name of the last folder (e.g., "nyt80_1")
    folder_name = os.path.basename(folder_path)
    
    # Get a list of all ZIP files in the folder
    zip_files = [file_name for file_name in os.listdir(folder_path) if file_name.endswith('.ZIP')]
    
    # Sort the ZIP files by the numeric part of the filenames (e.g., '_1.ZIP', '_2.ZIP', ...)
    zip_files.sort(key=lambda x: int(re.search(r'_(\d+)\.ZIP', x).group(1)))

    # Loop through all the sorted ZIP files
    for file_name in zip_files:
        zip_path = os.path.join(folder_path, file_name)
        print(f"Processing: {zip_path}")
        
        # Assuming extract_docx_from_zip is defined elsewhere
        docx_files = extract_docx_from_zip(zip_path)
            
        # For each .docx file in the ZIP archive, extract relevant data
        for docx_bytes in docx_files:
            doc = Document(docx_bytes)
            matched_data = inner_titles_extraction(doc, file_name)  # Pass folder_name to the extraction function
            all_data.extend(matched_data)
    
    # Create a DataFrame from the list of matched data
    df = pd.DataFrame(all_data)
    
    return df
```
